## **ESCALONAMENTO**

```python
from sklearn.preprocessing import StandardScaler
```

```python
esc = StandardScaler()
df_esc = esc.fit_transform(df)
```

```python
df_esc
```

## **K MEANS - PCA**

```python
df_esc.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
```

```python
df_pca = pca.fit_transform(df_esc)
```

```python
df_pca
```

```python
# razão das variáveis explicativas
pca.explained_variance_ratio_
```

```python
# Soma das variáveis explicativas
pca.explained_variance_ratio_.sum()
```

**Método Elbow**

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 5, max_iter = 300)
    kmeans.fit(df_pca)

    # Método inertia para gerar o wcss

    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(range(1, 11), wcss,marker='o',color='blue')
plt.title('Método Elbow')
plt.xlabel('Número de clusters')
plt.ylabel('WCSS');
```

**Agrupamento**

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 5, max_iter = 300)
```

```python
kmeans_fit = kmeans.fit(df_pca)
```

```python
# Coordenadas dos Centróides
centroides = kmeans_fit.cluster_centers_
centroides
```

```python
# Classificação dos dados
classificacao = kmeans_fit.labels_
classificacao
```

```python
classificacao[2]
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
graf1 = px.scatter(x = df_pca[:,0], y = df_pca[:,1], color=classificacao)
graf2 = px.scatter(x = centroides[:,0], y = centroides[:,1], size = [15, 15, 15, 15])
graf3 = go.Figure(data = graf1.data + graf2.data)
graf3.update_layout(width=800,height=500,title_text='Agrupamento K-Means com PCA')
graf3.update_xaxes(title = 'Componente 1')
graf3.update_yaxes(title = 'Componente 2')
graf3.show()
```
